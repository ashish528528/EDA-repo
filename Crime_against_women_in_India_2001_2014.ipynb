```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline

```

```python
file_path = '/content/crimes_against_women_2001-2014.csv'
```

```python
raw_df = pd.read_csv(file_path)
raw_df.head()
```

```python
#Dtoping unamed 0 column
raw_df.drop(['Unnamed: 0'], axis = 1,inplace=True)
```

72 unique state/ut not possible may be there is typing error.

```python
counts = raw_df['STATE/UT'].value_counts()
counts
```

```python
print(list(raw_df['STATE/UT']))
```

Seems its uppercase & lowercase and witespace problem

```python
#lets make entire state/ut column upper case
#and remove spaces between them
raw_df['STATE/UT'] = raw_df['STATE/UT'].apply(lambda x : x.upper())
raw_df["STATE/UT"] =  raw_df['STATE/UT'].apply(lambda x : x.replace(" ",""))
```

```python
raw_df['STATE/UT'].nunique()
```

Now seems little bit good 

```python
raw_df['Total crimes'] = raw_df.iloc[:,3:].sum(axis = 1)
```

```python
raw_df
```

```python
df = raw_df.copy()
```

```python
df.columns
```

```python
crimes = ['Rape', 'Kidnapping and Abduction',
       'Dowry Deaths', 'Assault on women with intent to outrage her modesty',
       'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
       'Importation of Girls', 'Total crimes']

df1=pd.DataFrame()
for i in crimes:
    df_crimes=df.groupby(['Year'])[i].sum()
    df1[i]=df_crimes
```
